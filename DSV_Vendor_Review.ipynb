```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
vendor_df = pd.read_excel(r'c:\Users\Phuong.Ngo\OneDrive - DSV\HP Working Folder\03. Reporting\02. Spend Analysis SOLUTIONS\Vendros masterdata (SOL).xlsx', sheet_name='VendorByGroup')
```

```python
vendor_df.head()
```

```python
#sns.displot(penguins, x="flipper_length_mm")
plt.figure(figsize=(12,6))
sns.set_context('paper', font_scale=1.4)
sns.set_style('darkgrid')
sns.distplot(vendor_df['Year of Established'], bins=5,color='darkblue')
```

```python
plt.figure(figsize=(12,6))
sns.set_context('paper', font_scale=1.5)
sns.set_style('darkgrid')
sns.countplot(x='Year of Established',data=vendor_df,color='darkblue')
plt.title('DSV Vendor Year established')
```

```python
plt.figure(figsize=(12,6))
sns.set_context('paper', font_scale=1.5)
sns.set_style('darkgrid')
plt.legend(loc=0)
sns.countplot(x='Branch',data=vendor_df, hue='Head Quarter')
plt.title('DSV Vendor by Branch & Head Quater')
```

```python
plt.figure(figsize=(12,6))
sns.set_context('paper', font_scale=1.5)
sns.set_style('darkgrid')
plt.legend(loc=0)
sns.countplot(x='Review List Check',data=vendor_df, hue='Branch')
plt.title('DSV Vendor by Review List')
```
